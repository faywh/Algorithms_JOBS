```python
import requests
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy import *
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
%matplotlib inline


#通过深交所获取股票历史数据
url="http://www.szse.cn/api/market/ssjjhq/getHistoryData?random=0.08048793408036281&cycleType=32&marketId=1&code=000058"
r=requests.get(url)
data=r.json()
```

```python
d = np.array(data['data']['picdowndata'])
d.shape
```

```python
data = np.array(d[:,1], dtype=int)
```

```python
data
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.plot(data)
```

```python
#数据归一化
scaler=MinMaxScaler(feature_range=(0,1))
dataset=scaler.fit_transform(data.reshape(len(data),1))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.plot(dataset)
```

```python
#从原始数据中分割出训练数据和测试数据
train_size=int(len(dataset)*0.618)
test_size=len(dataset)-train_size
train,test=dataset[:train_size,:],dataset[train_size:,:]
```

```python
#格式化数据便于输入
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(look_back,len(dataset)):
        a = dataset[i-look_back:i,0]                          
        dataX.append([a])
    Y=dataset[look_back:,0]
    return np.array(dataX), Y
```

```python
short_term=7 #设定短期周期为7天
train_X,train_Y=create_dataset(train,look_back=short_term)
test_X,test_Y=create_dataset(test,look_back=short_term)
origin_X,origin_Y=create_dataset(dataset,look_back=short_term)
```

```python
#使用序列模型训练
model=Sequential()
model.add(LSTM(50,activation='relu', input_shape=(train_X.shape[1],train_X.shape[2])))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam') 
```

```python
history = model.fit(train_X, train_Y, epochs=150, batch_size=72, validation_data=(test_X, test_Y))
```

```python
plt.plot(history.history["loss"],label="train")
plt.plot(history.history["val_loss"],label="test")
plt.legend()
plt.show()
```

```python
y_predicted=model.predict(test_X)
```

```python
plt.figure(figsize=(18, 6))
 
x_index=range(len(test_Y))
 
#反归一化，还原为原价
origin_Y=scaler.inverse_transform(np.array(test_Y).reshape((len(test_Y),1)))
y_predicted=scaler.inverse_transform(y_predicted)
 
plt.plot(x_index,origin_Y,label="true price",color="#008080")
plt.plot(x_index,y_predicted,label="predicted price",color="#ff6666")
 
/plt.legend()
```
